```python
import pandas as pd 
```

```python
# Missing Data Detection In Our Data Set

data = {
    "Name":[None,'Dayyan','Sudais','Khadim','Maaz','Arsalan','Moiz','Bilal'],
    "Age":[None,18,19,22,26,19,21,23],
    "Salary":[None,80000,90000,75000,100000,800000,65000,95000],
    "Performance Score":[None,60,50,50,95,65,70,95]
}

df = pd.DataFrame(data)
# True Means Data Is Missing And False Means Data Is Not Missing

print(df.isnull())
```

```python
# df.isnull().sum() Method 

data = {
    "Name":[None,'Dayyan','Sudais','Khadim','Maaz','Arsalan','Moiz','Bilal'],
    "Age":[None,18,19,22,26,19,21,23],
    "Salary":[None,80000,90000,75000,100000,800000,65000,95000],
    "Performance Score":[None,60,50,50,95,65,70,95]
}

df = pd.DataFrame(data)

# This method tells the total count of missing values in every column 

print(df.isnull().sum())
```

```python
# Handling Missing Data In Our Data Set:

data = {
    "Name":[None,'Dayyan','Sudais','Khadim','Maaz','Arsalan','Moiz','Bilal'],
    "Age":[None,18,19,22,26,19,21,23],
    "Salary":[None,80000,90000,75000,100000,800000,65000,95000],
    "Performance Score":[None,60,50,50,95,65,70,95]
}

df = pd.DataFrame(data)

df.dropna(inplace= True) # if i put axis=0 before inplace it will remove rows and axis=1 will remove columns 
print(df)
```

```python
# Adding Data In Place Of None Values 

data = {
    "Name":[None,'Dayyan','Sudais','Khadim','Maaz','Arsalan','Moiz','Bilal'],
    "Age":[None,18,19,22,26,19,21,23],
    "Salary":[None,80000,90000,75000,100000,800000,65000,95000],
    "Performance Score":[None,60,50,50,95,65,70,95]
}

df = pd.DataFrame(data)

df.fillna(0, inplace=True) # 0 will replace none values here 

print(df)
```

```python
# Adding Calculated Values Instead Of 0

data = {
    "Name":[None,'Dayyan','Sudais','Khadim','Maaz','Arsalan','Moiz','Bilal'],
    "Age":[None,18,19,22,26,19,21,23],
    "Salary":[None,80000,90000,75000,100000,800000,65000,95000],
    "Performance Score":[None,60,50,50,95,65,70,95]
}

df = pd.DataFrame(data)

df['Age'] = df['Age'].fillna(df['Age'].mean()) # Same Method For Other Columns Too

print(df)
```
