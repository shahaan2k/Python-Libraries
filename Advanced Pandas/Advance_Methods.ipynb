```python
import pandas as pd 
```

```python
# Adding Columns:

data = {
    "Name":['Shahaan','Dayyan','Sudais','Khadim','Maaz','Arsalan','Moiz','Bilal'],
    "Age":[20,18,19,22,26,19,21,23],
    "Salary":[50000,80000,90000,75000,100000,800000,65000,95000],
    "Performance Score":[55,60,50,50,95,65,70,95]
}

df = pd.DataFrame(data) 
print(df)

df["Bonus"] = df['Salary'] * 0.1 
print(df) 

# 2nd Method Using Insert() 
df.insert(0,"Employee Id", [10,20,30,40,50,60,70,80])
print(df)
```

```python
# Updating Values From Our Data Set:

data = {
    "Name":['Shahaan','Dayyan','Sudais','Khadim','Maaz','Arsalan','Moiz','Bilal'],
    "Age":[20,18,19,22,26,19,21,23],
    "Salary":[50000,80000,90000,75000,100000,800000,65000,95000],
    "Performance Score":[55,60,50,50,95,65,70,95]
}

df = pd.DataFrame(data)

df.loc[0,'Salary'] = 60000
print(df)

# Updating Multiple Values:
df['Salary'] = df['Salary'] * 1.05

print(df)
```

```python
# Removing Columns: 

data = {
    "Name":['Shahaan','Dayyan','Sudais','Khadim','Maaz','Arsalan','Moiz','Bilal'],
    "Age":[20,18,19,22,26,19,21,23],
    "Salary":[50000,80000,90000,75000,100000,800000,65000,95000],
    "Performance Score":[55,60,50,50,95,65,70,95]
}

df = pd.DataFrame(data)

df.drop(columns=['Performance Score'], inplace= True) # inplace=True means that it modifies the orignal data if it was false then it would create a new Data Set
print(f'Modified Data: \n {df}') 

# For Removing Multiple Columns You Simple Need To Put A Comma After Performance Score And Type The Name Of Other Column Inside The Square Brackets 
```
