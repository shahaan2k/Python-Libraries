```python
import pandas as pd 
```

```python
# Merging Data Sets 
```

```python
df_customers = pd.DataFrame({
    "CustomerId":[1,2,3],
    "Name":['Shahaan','Dayyan','Sudais']
})

df_orders = pd.DataFrame({
    "CustomerId":[1,2,4],
    "Order Amount":[100,250,350]
}) 

df_merged = pd.merge(df_customers, df_orders, on="CustomerId", how="inner")
print("Right Join")
print(df_merged) 
```

```python
# Concatenation:
```

```python
df_region1 = pd.DataFrame({
    "CustomerId":[1,2],
    "Name":['Shahaan','Dayyan']
})

df_region2 = pd.DataFrame({
    "CustomerId":[3,4],
    "Name":['Moiz','Arsalan']
}) 

df_concate = pd.concat([df_region1, df_region2], axis=1, ignore_index=True) 

print(df_concate)
```
