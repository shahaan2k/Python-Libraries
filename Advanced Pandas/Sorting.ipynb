```python
import pandas as pd 
```

```python
# Sorting 
```

```python
data = {
    "Name":['Shahaan', 'Dayyan', 'Khadim'],
    "Age":[20,18,24],
    "Salary":[10000,15000,20000]
}

df = pd.DataFrame(data) 

df.sort_values(by="Age", ascending=True, inplace= True) #Ascending true means in ascending order and false means descending order 

print(df)
```

```python
# Multiple Columns Sorting 

data = {
    "Name":['Shahaan', 'Dayyan', 'Khadim'],
    "Age":[20,18,24],
    "Salary":[10000,15000,20000]
}

df = pd.DataFrame(data) 

df.sort_values(by=["Age","Salary"], ascending=[True,False], inplace= True)

print(df)
```

```python
# Aggregating Data:

data = {
    "Name":['Shahaan', 'Dayyan', 'Khadim'],
    "Age":[20,18,24],
    "Salary":[10000,15000,20000]
}

df = pd.DataFrame(data) 

avg_salary = df['Salary'].mean() # .sum, .min, .max 

print(avg_salary) 
```

```python
# Grouping:

data = {
    "Name":['Shahaan', 'Dayyan', 'Khadim','Sudais', 'Moiz'],
    "Age":[20,18,24,20,24],
    "Salary":[10000,15000,20000,60000,75000]
}

df = pd.DataFrame(data) 
grouped = df.groupby('Age')['Salary'].sum()

print(grouped)
```

```python
# Multiple Columns Grouping 
```

```python
data = {
    "Name":['Shahaan', 'Dayyan', 'Khadim','Sudais', 'Moiz'],
    "Age":[20,18,24,20,24],
    "Salary":[10000,15000,20000,60000,75000]
}

df = pd.DataFrame(data) 
grouped = df.groupby(['Age','Name'])['Salary'].sum()

print(grouped)
```
