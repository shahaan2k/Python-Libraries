```python
import pandas as pd 
```

```python
# Interpolation Method:
# 1) Linear Method
# 2) Polynomial Method
# 3) Time Method 
```

```python
data = {
    "Name":['Shahaan','Dayyan','Sudais','Khadim','Maaz','Arsalan','Moiz','Bilal'],
    "Age":[20,18,None,22,26,19,21,23],
    "Salary":[None,80000,90000,75000,100000,800000,65000,95000],
    "Performance Score":[None,60,50,50,95,65,70,95]
}

df = pd.DataFrame(data)

df['Age'] = df['Age'].interpolate(method="linear")
print(df)
```

```python
data = {
    "Time":[1,2,3,4,5],
    "Value":[10,None,30,None,50]
} 

df = pd.DataFrame(data)

df['Value'] = df['Value'].interpolate(method="linear")
print(df)
```
