# Facet Grid

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
var = sns.load_dataset("tips").head(100)
var
```

```python
fg = sns.FacetGrid(var,col="day",hue="sex")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(var,col="day",hue="sex",palette="rocket")
fg.map(plt.bar,"total_bill","tip",edgecolor="black").add_legend()
plt.show()
```
