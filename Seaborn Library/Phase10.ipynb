# Factor Plot Or Catplot:

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
var = sns.load_dataset("tips").head(20)
var
```

```python
sns.catplot(
    x="size",
    y="tip",
    data = var,
    #kind="point", # "bar","box","strip","swarm","violin","boxen","count" 
    hue="sex",
    height=10
)
plt.show()
```
