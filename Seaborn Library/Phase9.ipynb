# Box Plot:

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(
    x="day",
    y="total_bill",
    data=var,
    hue="sex",
    palette="plasma",
    order=["Fri","Sun","Thur","Sat"], # Our Choice Order
    showmeans=True, # Shows The Mean
    meanprops={"marker":"+"},
    linewidth=2,
    #orient="h" for horizontal graph but x and y axis be in numeric value for it to be in horizontal position
)
plt.show()
```

```python
# Another Way Of Horizontal Graph

sns.set(style="whitegrid")
sns.boxplot(
    data=var,
    palette="plasma",
    showmeans=True, # Shows The Mean
    meanprops={"marker":"+"},
    linewidth=2,
    orient="h" 
)
plt.show()
```

```python
# Single Data Box Plot
sns.set(style="whitegrid")
sns.boxplot(
    x=var["total_bill"] # For horizontal x=var["total_bill"]
)
plt.show()
```
