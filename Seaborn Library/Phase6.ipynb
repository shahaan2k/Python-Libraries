# Violin Plot:

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd 
```

```python
var = sns.load_dataset("tips") 
var
```

```python
sns.violinplot(
    x="day",
    y="total_bill",
    data = var,
    hue="time",
    linewidth=1,
    palette="hsv"
)
plt.show()
```

```python
sns.violinplot(
    x="sex",
    y="total_bill",
    data = var,
    order=["Male", "Female"],
    palette=["Blue","Red"],
    inner="stick" # {"box","quartile","point","stick",None}
)
plt.show()
```

```python
sns.violinplot(
    x="day",
    y="total_bill",
    data = var,
    hue="sex",
    split=True,
    palette=["Blue","Red"],
    density_norm="width"
)
plt.show()
```

# Single Violin Plot:

```python
sns.violinplot(
    x="total_bill", # If I Want A Vertical Plot Just Use y="total_bill" instead of x
    data = var
)
plt.show()
```
