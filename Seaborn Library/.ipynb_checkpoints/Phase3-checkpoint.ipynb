# Scatter Plot:

```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
dt = sns.load_dataset("penguins").head(30)
dt
```

```python
m = {'Male':'*', 'Female':'^'}
sns.scatterplot(
    data= dt,
    y='bill_length_mm',
    x='bill_depth_mm',
    hue='sex',
    style='sex',
    size='sex',
    palette='rocket',
    sizes=(100,70),
    alpha=1,
    markers=m
)
plt.show()
```
