# Count Plot:

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd 
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.countplot(
    x="sex",
    hue="smoker",
    data=var,
    palette="afmhot"
)
plt.show()
```

```python
sns.barplot(x="sex", y="size", data=var)
plt.title("Bar Plot")
plt.show()
```
