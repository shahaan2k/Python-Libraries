# Line Plot

```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,5,6,7,1]

z = pd.DataFrame({"var":var, "var_1":var_1})

sns.lineplot(x="var",y="var_1", data=z)
 
plt.show()
```

```python
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,5,6,7,1]

sns.lineplot(x=var,y=var_1)
 
plt.show()
```

```python
# Loading A Data Set Into Our Jupyter From GitHub:

git = sns.load_dataset("penguins").head(20) 
git
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm", data=git, hue="sex", style="sex", 
             palette="Accent",markers=['o','>'],dashes=False, legend="full")

plt.title("Python")
plt.grid() 
plt.show() 
```

# Bar Plot

```python
var = sns.load_dataset("penguins").head(30)
var
```

```python
order_1 = ['Dream', 'Torgersen', 'Biscoe']
sns.barplot(
    x=var.island,
    y=var.bill_length_mm,
    hue=var.sex,
    order = order_1,
    hue_order=['Female', 'Male'],
    errorbar=('ci',100),
    n_boot=2
) # Or sns.barplot(x="island", y="bill_length_mm", data=var hue="sex")
plt.show()
```

```python
sns.barplot(
    x=var.bill_depth_mm,
    y=var.bill_length_mm,
    orient='h'
    
) 
plt.show() 
```

```python
sns.barplot(
    x='island',
    y='bill_length_mm',
    data=var,
    hue='sex',
    palette='rocket',
    saturation=15,
    errcolor='black',
    errwidth='2',
    capsize=0.5
) 
plt.show()
```

```python
sns.set(style='darkgrid')
sns.barplot(
    x='island',
    y='bill_length_mm',
    data=var,
    hue='sex',
    palette='rocket',
    saturation=15,
    errcolor='black',
    errwidth='2',
    capsize=0.5,
    dodge=False,
    alpha=0.5
) 
plt.show()
```
