# Pair Plot:

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns 
```

```python
var = sns.load_dataset("tips").head(20)
var
```

```python
sns.pairplot(
    var,
    hue="sex"
)
plt.show()
```

```python
sns.pairplot(
    var,
    vars=["total_bill","tip"],
    hue="sex",
    hue_order=["Female","Male"],
    palette="rocket"
)
plt.show()
```

```python
sns.pairplot(
    var,
    x_vars=["total_bill","tip"], # Same For Y Axis Too
    hue="sex",
    hue_order=["Female","Male"],
    palette="rocket",
    kind="kde"  # Type Of Graph # "scatter","reg","hist"
)
plt.show()
```
