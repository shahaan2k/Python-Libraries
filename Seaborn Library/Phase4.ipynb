# Heatmap:

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd  
import numpy as np
```

```python
var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var)
plt.show()
```

```python
dt = sns.load_dataset("anagrams").head(10)
dt.drop(columns=['attnr'], inplace=True)
dt
```

```python
var9 = np.linspace(1,10,10).reshape(2,5)
sns.heatmap(
    var9,
    vmin=0,
    vmax=10,
    cmap='PuOr',
    annot=True
)
```

```python
arr = np.array([
    ["a0","a1","a2","a3","a4"],
    ["b0","b1","b2","b3","b4",]
])

y = {"fontsize":15, "color":"black"}
sns.heatmap(
    var9, 
    vmin=0,
    vmax=10,
    cmap='Reds',
    annot=arr,
    annot_kws=y,
    fmt="s",
    linewidth=4,
    linecolor="black",
    cbar=False, # Will Remove The Color Bar
    xticklabels = False # Remove Labels 
)
plt.show()
```

```python
arr = np.array([
    ["a0","a1","a2","a3","a4"],
    ["b0","b1","b2","b3","b4",]
])

y = {"fontsize":15, "color":"black"}
v =sns.heatmap(
    var9, 
    vmin=0,
    vmax=10,
    cmap='Reds',
    annot=arr,
    annot_kws=y,
    fmt="s",
    linewidth=4,
    linecolor="black",
    cbar=False, # Will Remove The Color Bar
)
v.set(xlabel="Python", ylabel="WsCube")
sns.set(font_scale=3) # Font Scale Of X,Y Labels
plt.show()
```
