# Strip Plot:

```python
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
var = sns.load_dataset("tips").head(100)
var
```

```python
sns.stripplot(
    x="day",
    y="total_bill",
    data=var,
    hue="sex",
    palette="rocket",
    #linewidth=1,
    #edgecolor="r",
    #jitter=3,
    size=7,
    #marker="o"
)
plt.show()
```

```python
sns.stripplot(x=var["total_bill"]) # Same For y 
plt.show
```
