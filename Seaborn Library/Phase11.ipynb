# Styling Plots:

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd 
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.boxplot(
    y="total_bill",
    data=var
           )
plt.show()
```

```python
sns.set_style ("darkgrid")
sns.barplot(
    x="day",
    y="total_bill",
    data=var
           )
plt.show()
```

```python
sns.set_style ("whitegrid")
sns.barplot(
    x="day",
    y="total_bill",
    data=var
           )
sns.despine() # removes axis lines 
plt.show()
```

```python
# Figure Scaling
sns.set_style ("whitegrid")
plt.figure(figsize=(12,8))
sns.barplot(
    x="day",
    y="total_bill",
    data=var
           )
sns.despine() # removes axis lines 
plt.show()
```

```python
# Content Styling
sns.set_style ("whitegrid")
sns.set_context("poster",font_scale=0.8) # "paper"
sns.barplot(
    x="day",
    y="total_bill",
    data=var,
    palette="rocket",
    #legend=False,
    hue="day"
           )
sns.despine() # removes axis lines 
plt.show()
```
