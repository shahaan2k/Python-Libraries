```python
from sklearn.preprocessing import LabelEncoder 
```

```python
# Filling A Nan Value 
data = {
    'Name':["Shahaan","Dayyan","Khadim","Sudais","Moiz"],
    'Age':[21,None,44,23,None],
    'Salary':[50000,60000,70000,None,None]
}
df =pd.DataFrame(data)
print("Orignal Data Frame\n", df)

df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary']= df['Salary'].fillna(df['Salary'].mean())
print("Edited Data Frame\n",df)
```

# Label Encoding:

```python
df1 = pd.read_csv("Gender_Classification_Data.csv", encoding='latin1')
print("Orignal Data Set\n",df1)

df1_label = df1.copy() # Making A Safe Duplicate Of Data Set

le = LabelEncoder()
df1_label['Gender_Encoded']= le.fit_transform(df1_label['gender'])

print(" Edited Data Set\n Label Encoded Data")
print(df1_label[['gender','Gender_Encoded','height','weight','age']].head(10))
```

# Hot Encoding:

```python
data = {
    'Name':['Shahaan','Dayyan','Sudais','Khadim','Arsalan','Moiz','Bilal','Hammad','Khizar','Umer'],
    'City':['Islamabad','Islamabad','Peshawar','Charsadda','Lahore','Karachi','Skardu','Lahore','Islamabad','Abbotabad']
}

df = pd.DataFrame(data)
print(df)

df_encoded = pd.get_dummies(df,columns=['City'])
print('\n One Hot Encoded Data Of City')
df_encoded
```

# Standard Scaling And Min Max Scaling:

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split

data ={
    'Study Hours':[1,2,3,4,5,6],
    'Marks':[40,50,60,70,80,90]
}

df= pd.DataFrame(data)

# Standard Scaler 
standard_scaler = StandardScaler()
standard_scaled = standard_scaler.fit_transform(df)

print("Standard Scaler Output:")
print(pd.DataFrame(standard_scaled,columns=['Study Hours','Marks']))

# MinMax Scaling:

minmax_scaler = MinMaxScaler()
minmax_scaled = minmax_scaler.fit_transform(df)

print("\n MinMax Scaled Output:")
print(pd.DataFrame(minmax_scaled, columns=['Study Hours','Marks'])) 

# Train, Test And Split

x = df[['Study Hours']]
y = df[['Marks']]

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
print("x Training Data")
print(x_train)

print("x Testing Data")
print(x_test) 
```
