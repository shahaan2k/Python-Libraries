```python
# A Program That Predicts The Salary Of An Employee Based On Years Of Experience:
```

```python
from sklearn.linear_model import LinearRegression 

experience = [[1],[2],[3],[4],[5]]
salary = [75000, 150000, 200000, 250000, 350000]

model = LinearRegression()

model.fit(experience,salary) 

years = float(input("Enter Your Experience Period")) 

predicted_salary = model.predict([[years]])

print(f"BASED ON YOUR EXPERIENCE TIME {years} YOUR SALARY MAYBE AROUND {predicted_salary}") 
```
