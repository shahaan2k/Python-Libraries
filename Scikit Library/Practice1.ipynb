```python
import pandas as pd 
import seaborn as sns 
```

```python
var = sns.load_dataset("penguins").head(100)
df = pd.DataFrame(var)
```

```python
print("Orignal Data Set\n")
df
```

```python
df['bill_length_mm'] = df['bill_length_mm'].fillna(df['bill_length_mm'].mean())
df['bill_depth_mm'] = df['bill_depth_mm'].fillna(df['bill_depth_mm'].mean())
df['flipper_length_mm'] = df['flipper_length_mm'].fillna(df['flipper_length_mm'].mean())
df['body_mass_g'] = df['body_mass_g'].fillna(df['body_mass_g'].mean())
df['sex'] = df['sex'].fillna("Not Specified")
print("Edited Data Set:") 
df  # As You Can See The Difference Between Orignal Data Set And Edited Data Set Nan Values Are Edited Now But We Cannot Calculate The Mean Of A
    # Categorical Value Thats Why I Passed A String. 
```

```python
# Train, Test, Split, Standard Scaling And Cleaning The Data

data = {
    'Name':['Sahaan','Dayyan','Ayesha','Sudais','Javeria'],
    'Age':[21,18,None,19,21],
    'Class':[15,12,16,13,15],
    'Marks':[50,60,70,80,90],
    'Study Hours':[1,2,3,4,5],
    'Gender':['Male','Male','Female',None,'Female'],
}

df = pd.DataFrame(data)
df

# Step 1: Cleaning The Data:

df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Gender'] = df['Gender'].fillna('Not Specified')

print(df)

# Step 2: Label Encoding 

from sklearn.preprocessing import LabelEncoder

df_label = df.copy()
le = LabelEncoder()

df_label['Gender Encoder'] = le.fit_transform(df_label['Gender'])
print(df_label[['Name','Gender','Gender Encoder']])

# Step 3: Standard Scaling

from sklearn.preprocessing import StandardScaler, MinMaxScaler
numerical_columns = ['Age','Marks','Study Hours','Class']
standard_scaler = StandardScaler()
standard_scaled = standard_scaler.fit_transform(df[numerical_columns])

print("\n Standard Scaling Of DataSet:")
print(pd.DataFrame(standard_scaled,columns=['Age','Marks','Study Hours','Class']))

# Step 4:  Min Max Scaling 

minmax_scaling = MinMaxScaler()
minmax_scaled = minmax_scaling.fit_transform(df[numerical_columns])

print("\n MinMax Scaling")
print(pd.DataFrame(minmax_scaled,columns=['Age','Marks','Study Hours','Class']))

# Step 5 Train, Test, Split The Model:

from sklearn.model_selection import train_test_split

x =df [['Study Hours']]
y =df [['Marks']]

x_train, x_test, y_train, y_test = train_test_split(x,y,  test_size=0.2, random_state=42) 

print("X Training\n", x_train, "\n X Testing\n", x_test)
print("\n Y Training\n", y_train, "\n Y Testing\n", y_test)
```
