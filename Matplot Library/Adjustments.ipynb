```python
import matplotlib.pyplot as plt 
```

```python
x = [1 ,2, 3, 4]
y =[10, 25, 18, 30]

plt.subplot(1,2,1) # 1st Row , 2nd Column, 1st Subplot
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2) # 1st Row, 2nd Column, 2nd Subplot
plt.bar(x,y)
plt.title('Bar Chart')

plt.tight_layout() #Adjusts the spacing of the elements 
plt.show()
```

```python
# PROFESSIONAL WAY OF SUBPLOTS 
```

```python
# fig, ax= plt.subplots(nrows, ncols, figsize=(width, height))
fig, ax= plt.subplots(1,2, figsize=(10,5))

x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y)
ax[0].set_title('Line Plot')

ax[1].bar(x,y)
ax[1].set_title('Bar Chart')
plt.tight_layout() 

plt.show()
```

```python
# Way Of Customizing An Individual SubPlot
```

```python
fig, ax= plt.subplots(1,2, figsize=(10,5))

x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y, color = 'blue')
ax[0].set_title('Line Plot')
ax[0].grid(True)

ax[1].bar(x,y, color = 'green')
ax[1].set_title('Bar Chart')


fig.suptitle('Comparison Of Line And Bar Chart')
plt.tight_layout() 

plt.show()
```

```python
# How To Save A Figure 
```

```python
x = [1,2,3,4,5,6,7,8]
y = [10,0,10,20,30,0,0,0]

# Creating Plot
plt.plot(x,y, color='red', marker='o')
plt.title('Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.savefig('Line_Plot.png', dpi=300, bbox_inches='tight')
plt.show()
```
