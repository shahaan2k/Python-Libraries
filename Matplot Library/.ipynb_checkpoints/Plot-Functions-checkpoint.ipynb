```python
import matplotlib.pyplot as plt
```

```python
# plt.plot(x,y, color='color name' , linestyle='line_Style' , linewidth=value , marker='marker symbol' , label='label name') 

months = [1,2,3,4]
sales = [100,1500,1800,1200]

plt.plot(months,sales, color='black', linestyle='--', linewidth=1, marker='o', label='2026 sales data')
plt.xlabel("Months:")
plt.ylabel("No Of Sales:")
plt.title("Monthly Sales Data Of Unique:")
plt.legend(loc='upper left', fontsize=12)
plt.grid(color='black', linestyle=':', linewidth=1) 
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],['M1', 'M2', 'M3', 'M4'])

plt.show()
```

```python
# BAR CHARTS 
product = ['A', 'B', 'C', 'D']
sales = [500,800, 1400, 1900]

plt.bar(product,sales, color='Green', label='Sales 2026')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales Comparison')
plt.legend()
# BAR CHARTS CAN BE VERTICAL AND HORIZONTAL BOTH 
# plt.barh()

plt.show()
```

```python
# PIE CHARTS 
# plt.pie(values, labels=label_list, color=color-list,autopct='%1.1f&&')

regions = ['North','South','East','West']
revenue = [1000, 2000, 2500, 3000]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['red','green','blue','brown'])
plt.title('Revenue Contribution By Regions')

plt.show()
```

```python
# HISTOGRAMS

scores = [45,53,52,54,61,67,69,63,79,75,87,84,92,95,99,91,100]

plt.hist(scores, bins=[45,55,65,75,85,95,105], color='purple', edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('No Of Students')
plt.title('Students Score Comparison')
plt.show()
```

```python
# SCATTER PLOTTING 

hours_studied = [1,2,3,4,5,6,7,8]
exam_score = [50,55,60,65,70,75,80,85]

#plt.scatter(hours_studied,exam_score, color='red', marker='*', label='Student Data')

plt.xlabel('Studied Hours')
plt.ylabel('Score Range')
plt.title('Relationship Between Study Time And Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# MULTIPLE SCATTERING 

plt.scatter ([1,2,3],[50,55,60], color='yellow', label='A Class', marker='o')
plt.scatter ([1,2,3],[49,52,57], color='green', label='B Class', marker='^')

plt.xlabel('Studied Hours')
plt.ylabel('Score Range')
plt.title('Comparison Of Two Classes:')
plt.legend()
plt.grid(True)

plt.show()
```
