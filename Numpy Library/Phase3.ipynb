# Advance Operations With Business Examples:

```python
import numpy as np 
import matplotlib.pyplot as plt  
```

```python
# Data Structure: [resturant_id, 2022, 2023, 2024, 2025]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], # Paradise Burger 
    [2, 120000, 140000, 160000, 190000], # Beijing Bites 
    [3, 200000, 230000, 260000, 300000], # Pizza Hub
    [4, 180000, 210000, 240000, 270000], # Burger Point 
    [5, 160000, 185000, 205000, 230000]  # Chai Point  
])
```

```python
print("=== Zomato Sales Analysis ===") 
print("\n sales data shape:", sales_data.shape)
print("\n Sample Data For 1st Three Results\n", sales_data[:, 1:]) 

# TOTAL SALES PER YEAR 
# 1st Method
print(np.sum(sales_data, axis=0))
# 2nd Method
yearly_total = np.sum(sales_data[:, 1:], axis=0 )
print("Yearly Total Sales:",yearly_total)  

grand_total = np.sum(yearly_total)
print("Grand Total Of Sales:",grand_total)

# MINIMUM SALES PER RESTURANT 
min_sales = np.min(sales_data[:, 1:], axis=1)
print("Minimum Sales:",min_sales) 

# MAXIMUM SALES PER YEAR 
max_sales = np.max(sales_data[:, 1:], axis=0)
print("Maximum Sales:",max_sales)

# AVERAGE SALES PER RESTURANT 
avg_sales = np.mean(sales_data[:, 1:], axis=1) 
print("Average Sales:",avg_sales)

# CUMULATIVE SUM SALES 
cusum = np.cumsum(sales_data[:, 1:], axis=1) 
print("Cumulative Sum Of Sales:\n", cusum)

# MATPLOT LIBRARY 
plt.figure(figsize=(10,6))
plt.plot(np.mean(cusum, axis=0))
plt.title("Average Cumulative Sum:")
plt.xlabel("Years")
plt.ylabel("Sales") 
plt.grid(True)
plt.show() 

# VECTORS 
vector1 = np.array([2,4,6,8,10])
vector2 = np.array([1,5,7,9,11]) 
print("The Addition Of These Vectors:", vector1 + vector2) 
print("The Multiplication Of These Vectors:", vector1 * vector2) 
print("The Dot Product Of These Vectors", np.dot(vector1, vector2))

# Angle Of A Vector
angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1)* np.linalg.norm(vector2)))
print("The Angle Of Vectors:", angle)

restaurant_food = np.array(["biryani", "daal", "pizza", "burger"]) 
vectorized_up = np.vectorize(str.upper)
print("Upper Vectorize:", vectorized_up(restaurant_food)) 

# MONTHLY AVERAGE OF SALES DATA 
monthly_avg = sales_data[:, 1:]/12
print("The Monthly Average IS:\n", monthly_avg)
```
