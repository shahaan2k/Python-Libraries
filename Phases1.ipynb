# Numpy Array And Basics 
Learing About Numpy

```python
import numpy as np
```

# Creating Array From List

```python
arr_1d = np.array([1,2,3,4,5])
print("1D Array:",arr_1d) 

arr_2d = np.array([[4,7,9],[9,0,8]])
print("2D Array:",arr_2d)
```

# List Vs Numpy


```python
py_list = [1,2,3]
print("Python List Multiplication:", py_list * 2) 

np_array = np.array([1,2,3]) # Element Wise Multiplication 
print("Python Array Multiplication:", np_array * 2) 

import time

start = time.time()
py_list = [i*2 for i in range (1000000)]
print("\n List Operation Time: ", time.time() - start) 

start = time.time()
np_array = np.arange(1000000)*2
print("\n Array Operation Time:", time.time() - start)
```

# Creating Zeros Array

```python
zeros = np.zeros((3,4)) 
print("Zeros Array:\n",zeros)

ones = np.ones((5,5))
print("Ones Array: \n", ones)

# Now For A Array With Constant Number

full = np.full((2,3),7)
print("Full Array: \n",full)

# Now For Random Arrays 
random = np.random.random((2,2))
print("Random Array: \n",random)

# Now For Sequence Array
sequence = np.arange(0,11,2)
print("Sequence Array:\n", sequence)
```

# Vector, Matrix And Tensor 

```python
vector = np.array([1,2,3])
print("Vector:", vector)

matrix = np.array([[1,2,3],[4,5,6]])
print("Matrix:\n",matrix) 

tensor = np.array([[[2,3],[4,5],
                    [9,8],[8,1]]])
print("Tensor:\n",tensor)
```

# Array Properties:

```python
arr = np.array([[1,2,3],
                [6,7,8]]) 
print("Shape:", arr.shape)
print("Dimension:", arr.ndim)
print("Size:", arr.size)
print("Data Type", arr.dtype) 
```

# Array Reshaping

```python
arr = np.arange(12) 
print("Orignal Array",arr) 

# Reshape Array
reshape = arr.reshape((3,4))
print("Reshaped Array:", reshape)

# Flattened Array
flattened = reshape.flatten()
print("Flattened Array:", flattened)

# Raveled Array
raveled = reshape.ravel()
print("Raveled Array:", raveled) 

# Differnece In Ravel And FLatten <Flatten:Returns The Copy Of The Array> <Ravel: Returns The Orignal Form Of Array>

# Transpose
transpose = reshape.T # TRanspose: Converts Rows Into Columns And Columns Into Rows
print("Transposed Array:\n", transpose) 
```
