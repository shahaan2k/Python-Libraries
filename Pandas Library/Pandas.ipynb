```python
# Making A Data Frame And Saving Into Different Formats

data= {
    "Name":['Shahaan', 'Dayyan', 'Sudais', 'Khadim'],
    "Age":[20, 18, 19, 22],
    "City":['Peshawar', 'Peshawar', 'Peshawar', 'Charsadda']
}

df = pd.DataFrame(data)
print(df)
```

```python
# Saving In Different Formats:

df.to_csv("output.csv", index=False) # For CSV FIlE
print(df)
df.to_excel("output.xlsx", index=False) # FOR EXCEL FILE
print(df)
df.to_json("output.json", index=False) # FOR JSON FILE 
print(df)
```
