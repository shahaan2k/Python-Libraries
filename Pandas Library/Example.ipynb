```python
import pandas as pd 
```

```python
data = {
    "Name":['Shahaan','Dayyan','Sudais','Khadim','Maaz','Arsalan','Moiz','Bilal'],
    "Age":[20,18,19,22,26,19,21,23],
    "Salary":[50000,80000,90000,75000,100000,800000,65000,95000],
    "Performance Score":[55,60,50,50,95,65,70,95]
}

df = pd.DataFrame(data)

# Selecting A Single Column:
print("Name Column")
print(df["Name"])

# Multiple Columns:

subset = df[["Name", "Salary"]]
print(subset)
```

```python
# Filtering Rows 
# Single Condition Filtering:

data = {
    "Name":['Shahaan','Dayyan','Sudais','Khadim','Maaz','Arsalan','Moiz','Bilal'],
    "Age":[20,33,36,22,26,19,21,23],
    "Salary":[50000,80000,90000,75000,100000,800000,65000,95000],
    "Performance Score":[55,60,50,50,95,65,70,95]
}

df = pd.DataFrame(data)

high_salary = df[df["Salary"]>50000]
print(f"Employee With Salary > 50,000\n {high_salary}")

# Multiple Condition Filtering:
filtered = df[(df['Age']>=30) & (df['Salary']>=65000)]
print(f"Employer Older Than 30 And Salary >=60,000 \n {filtered}")

# Using OR Condition
condition = df[(df['Age'] >30) | (df['Performance Score']>60)]
print(f"Employes Whith Age > 30 Or Performance Score >60\n {condition}") # Evalutaes The Program If Even One Condition Is True
```
